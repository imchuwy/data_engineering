```python
df = pd.read_csv('orders.csv') #from csv
# pd.read_excel('data.xlxs') from excel
# pd.DataFrame('data_dict') from another data dictionary
df.head()
```

```python
#df.tail() # Last 5 rows
#df.info() # general info
#df.describe() # Statistics
#df.columns # List columns
```

```python
#Index column
#df["Country"] # Indexes by column name. We can wrap in "set" for unique value, "len" for count
df[["Country", "Product"]] # Two square brackets needed to show in 2d dataframe
```

```python
# Index on row. [row][column]

df.iloc[0]["Country"]
```

Filtering

```python
#This filters the dataframe to show rows where quantity = 2
#df["Quantity"] == 2 This filters the row for boolean T/F
#df[df["Quantity"] == 2] #This then uses the above and decides whether to show that row
df.loc[df["Quantity"] == 2] # This uses label based indexer. Above is boolean. 
```

```python
#This filters the dataframe on two rows
df[(df["Category"] == "Electronics") & (df["Country"] == "USA")] # This filters for Electronics AND USA

# & is and, | is or
```

```python
#String filters. also endswith
#df[df["CustomerName"].str.startswith("A")]

# isin. ~ before the wrapped df (2nd df) will show those not in. 
df[~df["Country"].isin(["USA", "Sweden", "Brazil"])]
```

```python
# Update a row

df.loc[df["CustomerName"] == "Anna Ivanova", "Product"] = "Monitor" # Find the row where Customername = ... Then make product "Monitor"
df.loc[df["CustomerName"] == "Anna Ivanova"]
```

```python
# To change a column
df["Country"] = df["Country"].str.upper()
df["Country"]
```

```python
df = df.drop(39) # drop row 39
df.dropna() # Return a Dataframe with drop rows with NA's
df.fillna({"OrderID": 0}, inplace=True) # Modify NA's in OrderID to 0
```

```python
#df["Country"].value_counts() #Count of distinct values
df.groupby("Country")["Price"].sum()
df.sort_value("Price", ascending=False)
```

```python
df.to_csv("new_file.csv", index=False)
```
