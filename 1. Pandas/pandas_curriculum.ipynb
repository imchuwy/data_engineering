# Pandas mini-curriculum
## Goals
- Load CSV/Excel/JSON into DataFrames; inspect shape/info/types
- Indexing, filtering, boolean masks, query strings
- Sorting, dropping, renaming, handling missing values
- GroupBy/agg, pivot tables, merge/join/concat
- Date/time handling and categorical data
- Exporting cleaned data

## Suggested exercises
- - Download two small CSVs and explore `df.head()`, `df.info()`, `df.describe()`
- - Write a function that cleans column names and fills missing values
- - Group sales by category and plot top 5
