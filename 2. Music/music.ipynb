```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
#Import the dataset and assign to a dataframe
music_data = pd.read_csv('music.csv')

#Shape the data for training
X = music_data.drop(columns=['genre']) #We drop this column. Capital X by convention for input set
y = music_data['genre'] #We select only column genre. Lowercase y as convention for output set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#Model prediction for some values
model = DecisionTreeClassifier()

#Fitting the model
model.fit(X, y) #(Input, Output)
model.fit(X_train, y_train)
predictions = model.predict([[21,1] , [22,0] , [30,1]]) #Here we ask the model to predict the music prefences of a 21 year old male. And a 22 year old female.
```

```python
#Test for model accuracy
model = DecisionTreeClassifier()

#Fitting the model
model.fit(X, y) #(Input, Output)
model.fit(X_train, y_train)
predictions = model.predict(X_test) #Here we ask the model to predict the music prefences of a 21 year old male. And a 22 year old female.
score = accuracy_score(y_test, predictions)
score
```
